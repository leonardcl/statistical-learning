## Import

```python
from sklearn.datasets import make_moons
from matplotlib import pyplot
from pandas import DataFrame
from keras.utils.np_utils import to_categorical
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import numpy as np
import seaborn as sns
from sklearn import metrics
sns.set_theme()
```

## Generate Train Dataset

```python
# generate 2d classification dataset
train_X, train_y = make_moons(n_samples=300, noise=0.15, random_state=2801)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=train_X[:,0], y=train_X[:,1], label=train_y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

## Generate Test Dataset

```python
# generate 2d classification dataset
test_X, test_y = make_moons(n_samples=100, noise=0.2, random_state=3007)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=test_X[:,0], y=test_X[:,1], label=test_y))
colors = {0:'red', 1:'blue'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.show()
```

## Train With 1 Neuron Model (Model 1)

```python
model_1 = Sequential()

# model_1.add(Dense(units=10, activation="swish", input_dim=2))
# model_1.add(Dense(units=10, activation="relu"))
model_1.add(Dense(units=1, activation="sigmoid"))

model_1.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

## Train With 2-2-1 Neuron Model (Model 2)

```python
model_2 = Sequential()

model_2.add(Dense(units=2, activation="swish", input_dim=2))
model_2.add(Dense(units=2, activation="relu"))
model_2.add(Dense(units=1, activation="sigmoid"))

model_2.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

## Train With 10-10-1 Neuron Model (Model 3)

```python
model_3 = Sequential()

model_3.add(Dense(units=10, activation="swish", input_dim=2))
model_3.add(Dense(units=10, activation="relu"))
model_3.add(Dense(units=1, activation="sigmoid"))

model_3.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

## Fit The Model

```python
hist_1 = model_1.fit(train_X, train_y, batch_size=5, epochs=100, verbose=1)
print(model_1.predict(train_X))
results_1 = model_1.predict(test_X)
```

```python
hist_2 = model_2.fit(train_X, train_y, batch_size=5, epochs=100, verbose=1)
print(model_2.predict(train_X))
results_2 = model_2.predict(test_X)
```

```python
hist_3 = model_3.fit(train_X, train_y, batch_size=5, epochs=100, verbose=1)
print(model_3.predict(train_X))
results_3 = model_3.predict(test_X)
```
